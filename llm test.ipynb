```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna(),cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing values")
```

```python
import numpy as np

books["missing_dis"] = np.where(books["description"].isna(), 0, 1)
books["age_of_books"] = 2024 - books["published_year"]
```

```python
columns_of_interest = ["num_pages","age_of_books","missing_dis","average_rating"]

corretion_matrix = books[columns_of_interest].corr(method="spearman")

sns.set_theme(style="whitegrid")
plt.figure(figsize=(10,10))
heatmap = sns.heatmap(corretion_matrix, annot=True, cmap="coolwarm",fmt=".2f",cbar_kws={"label": "Correlation"})
heatmap.set_xlabel("Columns")
heatmap.set_ylabel("Missing values")
plt.show()
```

```python
books_missing = books[~(books['description'].isna())&~(books['num_pages'].isna())&~(books['average_rating'].isna())&~(books['published_year'].isna())]
```

```python
books_missing
```

```python
books_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
books_missing["words_in_descryption"] = books_missing["description"].str.split().str.len()
```

```python
books_missing
```

```python
book_missing_25_words = books_missing[books_missing["words_in_descryption"] >= 25]
```

```python
book_missing_25_words
```

```python
book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],
             book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
book_missing_25_words
```

```python
(
    book_missing_25_words
    .drop(["subtitle", "missing_dis", "age_of_books", "words_in_descryption"], axis=1)
    .to_csv("books_cleaned.csv", index = False)
)
```
